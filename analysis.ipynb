# Visualizing the Data #
# Finding timestamp #

```python
# to see the opening price
# We can see the highest opening price was in the end of januari

tsla ['Open'].plot(figsize=(14,4),title='Visualizing for opening price')
plt.ylabel('Opening Price')
plt.xlabel('Datetime')
plt.show()
```

```python
# 26 jan 2021 was the highest price
# loc function for location 
# idxmax() function is used to get the row label of the maximum value

tsla.loc[tsla['Open'].idxmax()].name
```

```python
# Daily volume of traded stock
#we can see the big spikes of trades in 4 feb 2020

tsla['Volume'].plot(label='Volume',figsize=(14,4),title='Visualizing for traded volume')
plt.legend()
plt.ylabel('Traded Volume')
plt.xlabel('Datetime')
plt.show()
```

```python
# idxmax() function is used to get the row label of the maximum value

tsla.loc[tsla['Volume'].idxmax()].name
```

```python
# visualing of total traded and the highest trade is in 18 dec 2020

tsla['Total Traded'].plot(figsize=(14,4),title='Visualizing for total traded')
plt.ylabel('Total Traded')
plt.xlabel('Datetime')
plt.show()
```

```python
# timestamp for the highest trade values

tsla.loc[tsla['Total Traded'].idxmax()].name
```

```python
#When using the history() function each datetime index in the returned DataFrame object has the timestamp 00:00:00.
#see daily opening and closing price

tsla = yf.Ticker('TSLA').history(period='7d', interval='1h')[['Open', 'Close']]
tsla.head(10)
```

```python
#calculation for daily returns
#shift() function for scalar parameter over the desire axis 

tsla['Returns'] = (tsla['Close'] / tsla['Close'].shift(1) ) - 1 
tsla.head()
```

```python
# a histogram showing same hits of the same values

tsla['Returns'].hist(bins=80, figsize=(14, 5))
plt.ylabel('Frequency')
plt.xlabel('Returned Values')
```
